Encoding the target variable using hasing and using target encoder to encode the categroical variables. And then performing scaling using MinMax scaler.

```python
from sklearn.feature_extraction import FeatureHasher
from category_encoders import TargetEncoder
from sklearn.preprocessing import MinMaxScaler


track_names = df['track_name'].tolist()
hasher = FeatureHasher(n_features=1, input_type='string')
hashed_track_name = hasher.fit_transform([[track_name] for track_name in track_names]).toarray()

df['hashed_track_name'] = hashed_track_name

encoder = TargetEncoder(cols=string_columns)
df[string_columns] = encoder.fit_transform(df[string_columns], df['hashed_track_name'])


encoded_df = df.drop(columns=['track_name'])

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(encoded_df)
```

Visualization of a sample of the features

Calculating similarity matrix using cosine similarity and euclidean distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity_matrix = cosine_similarity(df_scaled)

def recommend_similar_tracks_cosine(track_name, similarity_matrix, k=5):
    track_index = df.index[df['track_name'] == track_name].tolist()[0]
    
    track_similarities = similarity_matrix[track_index]

    similar_track_indices = track_similarities.argsort()[::-1][1:k+1]  

    similar_tracks = df.iloc[similar_track_indices]['track_name'].tolist()
    
    return similar_tracks

similar_tracks = recommend_similar_tracks_cosine('Dance All Night', cosine_similarity_matrix)
print(similar_tracks)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

euclidean_distances_similarity_matrix = euclidean_distances(df_scaled)

def recommend_similar_tracks_euclidean(track_name, similarity_matrix, k=5):

    track_index = df.index[df['track_name'] == track_name].tolist()[0]

    track_similarities = similarity_matrix[track_index]

    similar_track_indices = track_similarities.argsort()[1:k+1] 
    
    similar_tracks = df.iloc[similar_track_indices]['track_name'].tolist()
    
    return similar_tracks

similar_tracks = recommend_similar_tracks_euclidean('Dance All Night', euclidean_distances_similarity_matrix)
print(similar_tracks)
```

Finding the best parameters for kmeans.

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_clusters': list(range(2, 24)),
    'init': ['k-means++', 'random'],
    'n_init': [10, 20, 30],
    'max_iter': [100, 200, 300],
    'tol': [1e-3, 1e-4, 1e-5] 
}

kmeans = KMeans()
random_search = RandomizedSearchCV(kmeans, param_distributions=param_grid, n_iter=10, random_state=2)

random_search.fit(X_df1_scaled)
best_kmeans = random_search.best_estimator_
print("Best parameters found:", random_search.best_params_)
print("Best score found:", random_search.best_score_)
```

```python
from sklearn.cluster import KMeans


best_params = {
    'tol': 0.00001,
    'n_init': 10,
    'n_clusters': 23,
    'max_iter': 300,
    'init': 'k-means++'
}

best_kmeans = KMeans(**best_params, random_state=42)
best_kmeans.fit(X_df1_scaled)
cluster_labels = best_kmeans.predict(X_df1_scaled)
```

```python
danceability_index = X_df1.columns.get_loc('danceability')
energy_index = X_df1.columns.get_loc('energy')

# Visualizing KMeans clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_df1_scaled[:, danceability_index], y=X_df1_scaled[:, energy_index], hue=cluster_labels, palette='viridis', legend='full')
plt.title('KMeans Clustering')
plt.xlabel('danceability')
plt.ylabel('energy_index')
plt.show()
```

```python
unique_values, counts = np.unique(cluster_labels, return_counts=True)

for value, count in zip(unique_values, counts):
    print(f"Cluster {value}: {count} occurrences")

print("Total number of clusters:", len(unique_values))
```

Finding the optimal min_samples for DBSCAN

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score

min_samples_range = range(5, 100) 
silhouette_scores = []
calinski_harabasz_scores = []
for min_samples in min_samples_range:
    dbscan_model = DBSCAN(eps=0.5, min_samples=min_samples)
    dbscan_model.fit(X_df1_scaled)
    
    silhouette = silhouette_score(X_df1_scaled, dbscan_model.labels_)
    silhouette_scores.append(silhouette)
    
    calinski_harabasz = calinski_harabasz_score(X_df1_scaled, dbscan_model.labels_)
    calinski_harabasz_scores.append(calinski_harabasz)

plt.figure(figsize=(10, 6))
plt.plot(min_samples_range, silhouette_scores, marker='o', label='Silhouette Score', color='blue')
plt.title('Silhouette Score vs. Min Samples (DBSCAN)')
plt.xlabel('Min Samples')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(min_samples_range, calinski_harabasz_scores, marker='o', label='Calinski-Harabasz Index', color='green')
plt.title('Calinski-Harabasz Index vs. Min Samples (DBSCAN)')
plt.xlabel('Min Samples')
plt.ylabel('Calinski-Harabasz Index')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def get_similar_songs_dbscan(song_name, df, model, scaled_data, top_n=5):
    if song_name in df['track_name'].values:
        song_index = df[df['track_name'] == song_name].index[0]
        if song_index < len(scaled_data):
            # Extracting cluster label of the input song
            song_cluster_label = model.labels_[song_index]
            cluster_indices = np.where(model.labels_ == song_cluster_label)[0]
            cluster_df = df.iloc[cluster_indices].drop(song_index)
            
            # Calculating cosine similarity within the cluster
            cosine_similarities = cosine_similarity(scaled_data[song_index].reshape(1, -1), scaled_data[cluster_indices]).flatten()
            
            # Sorting songs based on cosine similarity
            similar_songs_indices = np.argsort(cosine_similarities)[::-1][:top_n]
            similar_songs = cluster_df.iloc[similar_songs_indices]
            
            return similar_songs
        else:
            print("Song index is out of bounds.")
            return None
    else:
        print("Song name not found in DataFrame.")
        return None

song_name = "The Sound"
similar_songs_dbscan = get_similar_songs_dbscan(song_name, data, dbscan_model, X_df1_scaled)
if similar_songs_dbscan is not None:
    print("Recommended songs similar to", song_name)
    for index, row in similar_songs_dbscan.iterrows():
        print("Track Name:", row['track_name'])
        print("Artist Name:", row['track_artist'])
        print("Album Name:", row['track_album_name'])
        print()
```
