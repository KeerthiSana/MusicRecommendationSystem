```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the data

```python
data = pd.read_csv('/kaggle/input/30000-spotify-songs/spotify_songs.csv')
data
```

Checking for null values.

```python
data.isnull().sum()
```

Removing the null values

```python
df=data.dropna()
df.isnull().sum()
```

Removing duplicate rows based on the track_name(target variable).

```python
df = df.drop_duplicates(subset='track_name', keep='first')
df.info()
```

```python
df = df.drop(['track_id','playlist_id','track_album_id','duration_ms','track_album_release_date'], axis = 1)
df.info()
```

```python
string_columns = ['playlist_name','playlist_genre','playlist_subgenre', 'track_artist', 'track_album_name']
numerical_columns = ['track_popularity','danceability', 'energy','key','loudness','mode','speechiness', 'acousticness', 'instrumentalness','liveness','valence', 'tempo']
```

```python
df1=df.copy()
```

```python
df.reset_index(drop=True, inplace=True)
```

Encoding the target variable using hasing and using target encoder to encode the categroical variables. And then performing scaling using MinMax scaler.

```python
from sklearn.feature_extraction import FeatureHasher
from category_encoders import TargetEncoder
from sklearn.preprocessing import MinMaxScaler


track_names = df['track_name'].tolist()
hasher = FeatureHasher(n_features=1, input_type='string')
hashed_track_name = hasher.fit_transform([[track_name] for track_name in track_names]).toarray()

df['hashed_track_name'] = hashed_track_name

encoder = TargetEncoder(cols=string_columns)
df[string_columns] = encoder.fit_transform(df[string_columns], df['hashed_track_name'])


numeric_df = df.drop(columns=['track_name'])

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(numeric_df)
```

Calculating similarity matrix using cosine similarity and euclidean distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(df_scaled)

def recommend_similar_tracks_cosine(track_name, similarity_matrix, k=5):
    track_index = df.index[df['track_name'] == track_name].tolist()[0]
    
    track_similarities = similarity_matrix[track_index]

    similar_track_indices = track_similarities.argsort()[::-1][1:k+1]  

    similar_tracks = df.iloc[similar_track_indices]['track_name'].tolist()
    
    return similar_tracks

similar_tracks = recommend_similar_tracks_cosine('Dance All Night', similarity_matrix)
print(similar_tracks)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

similarity_matrix = euclidean_distances(df_scaled)

def recommend_similar_tracks_euclidean(track_name, similarity_matrix, k=5):

    track_index = df.index[df['track_name'] == track_name].tolist()[0]

    track_similarities = similarity_matrix[track_index]

    similar_track_indices = track_similarities.argsort()[1:k+1]  # Exclude the track itself
    
    similar_tracks = df.iloc[similar_track_indices]['track_name'].tolist()
    
    return similar_tracks

similar_tracks = recommend_similar_tracks_euclidean('Dance All Night', similarity_matrix)
print(similar_tracks)
```

Finding the best parameters for kmeans.

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_clusters': list(range(2, 24)),
    'init': ['k-means++', 'random'],
    'n_init': [10, 20, 30],
    'max_iter': [100, 200, 300],
    'tol': [1e-3, 1e-4, 1e-5] 
}

kmeans = KMeans()
random_search = RandomizedSearchCV(kmeans, param_distributions=param_grid, n_iter=10, random_state=2)

random_search.fit(X_df1_scaled)
best_kmeans = random_search.best_estimator_
print("Best parameters found:", random_search.best_params_)
print("Best score found:", random_search.best_score_)
```
